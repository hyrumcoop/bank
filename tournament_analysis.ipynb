```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('./results.csv')
df['winPercent'] = df['placementIn1'] / df['games_played']
```

```python
roundsof10 = df[df['roundsPerGame'] == 10]
roundsof15 = df[df['roundsPerGame'] == 15]
roundsof20 = df[df['roundsPerGame'] == 20]
```

## Round of 10 Analysis

```python
sns.barplot(roundsof10, x="playerName", y="winPercent", hue="playersPerTable")
plt.xticks(rotation=90)
plt.title('Wins by Player Type in Games of 10 Rounds')
plt.xlabel('Player Type')
plt.ylabel('Times Placed First')
plt.savefig('./charts/Wins10Rounds.png', format="png", bbox_inches="tight")
```
